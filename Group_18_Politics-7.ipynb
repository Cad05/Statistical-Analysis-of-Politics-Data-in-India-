3.3) Setting up the Age for our dataset

Lets check the Summary of the Column: Age for all of our datasets. 

```r
summary(data2004$Age)
summary(data2009$Age)
summary(data2014$Age)
summary(data2019$Age)
```

Technically the minimum value for Age for any candidate cannot be 0 also for our analysis we are not considering candidates whose age is greater than 80. So we are cleaning the Age column by deleting those values of Age where , Age is less that 18 and greater than 80

```r
data2004 <- data2004[-c(which(data2004$Age > 80 | data2004$Age < 18)), ]
data2009 <- data2009[-c(which(data2009$Age > 80 | data2009$Age < 18)), ]
data2014 <- data2004[-c(which(data2014$Age > 80 | data2014$Age < 18)), ]
data2019 <- data2004[-c(which(data2019$Age > 80 | data2019$Age < 18)), ]
```
